```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos de entrada (x) y salida (y)

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 3, 6, 7, 10])

# Crear la matriz de dise√±o X con una columna de 1's para el intercepto
# A gregar una columnas de 1's para el intercepto

X = np.vstack( [np.ones_like(x),x] ).T
print(f"Matriz de dise√±o X: {X}")

# Calcular los coeficientes beta usando la formula (X^T X)-1 X^T y

# 1. Calcular X^T
    
XT = X.T
print(f"X.T: {XT}")
    
XT_X = np.dot(XT, X)
print(f"XT_X: {XT_X}")
    
# 2. Calcular la inversa (X^T X) 
    
XT_XI = np.linalg.inv(XT_X)
print(f"Matriz inversa: {XT_XI}")
    
# 3. Calcular los coeficientes beta
    
beta = XT_XI @ X.T @ y
print(f"Beta: f.16 {beta}")

#Los coeficientes beta representan [b, m] para la ecuaci√≥n y = m * x + b

intercepto = -1.2 # b (intercepto)
pendiente = 2.2  # m (pendiente)

print(f"Coeficiente (intercepto) b: {intercepto:.20f}")
print(f"Pendiente m: {pendiente:.16f}")

# Predicciones con la recta de regresi√≥n

y_pred = np.dot( X , beta) # Y = m * X + b

# Graficar los datos orginales (puntos)

plt.scatter(x, y, color="blue", label="Datos", marker="o")

# Graficar la linea de regresi√≥n lineal
plt.plot(x, y_pred, color="red", label=f"Regesi√≥n Lineal: y = {pendiente:.2f}x + {intercepto:.2f}", linewidth=2)

# Etiqueta y t√≠tulo

plt.xlabel("Horas de estudio")
plt.ylabel("Calificaci√≥n")
plt.title("Regresi√≥n lineal: Horas de Estudio vs Calificaci√≥n")

# Mostrar leyenda

plt.legend()

#Mostrar la gr√°fica 

plt.grid(True)
plt.show()
```
